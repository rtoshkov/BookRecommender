```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import pandas as pd
books = pd.read_csv('books_cleaned.csv')
```

```python
books
```

```python
books["tagged_description"].to_csv("tagged_descriptions.csv", index=False, header=False)
```

```python
raw_documents = TextLoader("tagged_descriptions.csv", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
db_books = Chroma.from_documents(documents, embedding=OpenAIEmbeddings())
```

```python
query = "A book to teach about nature"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0][1:].strip())]
```

```python
def retrieve_semantiv_recommendations (query: str, top_k: int = 10) -> pd.DataFrame:
    recs = db_books.similarity_search(query=query, k=50)
    book_list = []
    for i in range(0,len(recs)):
        book_list += [int(recs[i].page_content.strip('"').split()[0])]
    return books[books["isbn13"].isin(book_list)].head(top_k)
```

```python
retrieve_semantiv_recommendations("A book to teach kids about nature")
```
